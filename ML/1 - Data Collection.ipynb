```python
f1 = open('list_of_all_coins.txt')
f2 = open('list_coins_available_blockchain_data.txt')

list_of_all_coins = ast.literal_eval(f1.read())
list_coins_available_blockchain_data = ast.literal_eval(f2.read())
```

```python
currency_symbol = 'BTC'
currency_id = 1182
limit_of_request = 1000
```

```python
#demo_dict = ["BTC": {"id": '123', "data": {"11923": {"price": 20, "volume": 3, "twitter": 1}}]
raw_data = []
```

```python
currencies = list_coins_available_blockchain_data#[{"name": "BTC", "id": 1182}, {"name": "ETH", "id": 7605}]
```

```python
for currency in range(0, 1):
    print(currencies[currency]["name"], )

    request = requests.get(f'https://min-api.cryptocompare.com/data/v2/histoday?fsym={currencies[currency]["name"]}&tsym=USD&limit={limit_of_request}')
    loaded_data = json.loads(request.content)

    if ("Data" in loaded_data["Data"]):
        loaded_data = loaded_data["Data"]["Data"]

        data_list = []

        for i in range(0, len(loaded_data)):
            tmp_dict = {}
            tmp_dict["time"] = loaded_data[i]["time"]
            tmp_dict["price"] = loaded_data[i]["close"]
            tmp_dict["volume_from"] = loaded_data[i]["volumefrom"]
            tmp_dict["volume_to"] = loaded_data[i]["volumeto"]

            data_list.append(tmp_dict)

        request_social = requests.get(f'https://min-api.cryptocompare.com/data/social/coin/histo/day?coin_id={currencies[currency]["id"]}&limit={limit_of_request}&api_key=d8b661610aa93dcd2adf734344ae50304d08d0fc02f9ff51ba7cbc4e66f94533')
        loaded_data_social = json.loads(request_social.content)
        loaded_data_social = loaded_data_social["Data"]

        for i in range(0, len(loaded_data_social)):
            for j in range(0, len(data_list)):
                if (data_list[j]["time"] == loaded_data_social[i]["time"]):
                    data_list[j]["comments"] = loaded_data_social[i]["comments"]
                    data_list[j]["points"] = loaded_data_social[i]["points"]
                    data_list[j]["followers"] = loaded_data_social[i]["followers"]
                    data_list[j]["fb_likes"] = loaded_data_social[i]["fb_likes"]
                    data_list[j]["twitter_followers"] = loaded_data_social[i]["twitter_followers"]
                    data_list[j]["twitter_statuses"] = loaded_data_social[i]["twitter_statuses"]
                    data_list[j]["reddit_subscribers"] = loaded_data_social[i]["reddit_subscribers"]
                    data_list[j]["reddit_active_users"] = loaded_data_social[i]["reddit_active_users"]
                    data_list[j]["reddit_posts_per_day"] = loaded_data_social[i]["reddit_posts_per_day"]
                    data_list[j]["reddit_comments_per_day"] = loaded_data_social[i]["reddit_comments_per_day"]



        request_blockchain = requests.get(f'https://min-api.cryptocompare.com/data/blockchain/histo/day?fsym={currencies[currency]["name"]}&limit={limit_of_request}&api_key=d8b661610aa93dcd2adf734344ae50304d08d0fc02f9ff51ba7cbc4e66f94533')
        loaded_data_blockchain = json.loads(request_blockchain.content)
        loaded_data_blockchain = loaded_data_blockchain["Data"]["Data"]

        for i in range(0, len(loaded_data_blockchain)):
            for j in range(0, len(data_list)):
                if (data_list[j]["time"] == loaded_data_blockchain[i]["time"]):
                    data_list[j]["new_addresses"] = loaded_data_blockchain[i]["new_addresses"]
                    data_list[j]["transaction_count"] = loaded_data_blockchain[i]["transaction_count"]
                    data_list[j]["large_transaction_count"] = loaded_data_blockchain[i]["large_transaction_count"]
                    data_list[j]["average_transaction_value"] = loaded_data_blockchain[i]["average_transaction_value"]
                    data_list[j]["unique_addresses_all_time"] = loaded_data_blockchain[i]["unique_addresses_all_time"]
                    data_list[j]["current_supply"] = loaded_data_blockchain[i]["current_supply"]
                    if ("block_time" in data_list[j]):
                        data_list[j]["block_time"] = loaded_data_blockchain[i]["block_time"]
                    data_list[j]["new_addresses"] = loaded_data_blockchain[i]["new_addresses"]
                    if ("block_size" in data_list[j]):
                        data_list[j]["block_size"] = loaded_data_blockchain[i]["block_size"]
                    if ("hashrate" in data_list[j]):
                        data_list[j]["hashrate"] = loaded_data_blockchain[i]["hashrate"]
                    if ("block_height" in data_list[j]):
                        data_list[j]["block_height"] = loaded_data_blockchain[i]["block_height"]
                    if ("zero_balance_addresses_all_time" in data_list[j]):
                        data_list[j]["zero_balance_addresses_all_time"] = loaded_data_blockchain[i]["zero_balance_addresses_all_time"]
                    if ("transaction_count_all_time" in data_list[j]):
                        data_list[j]["transaction_count_all_time"] = loaded_data_blockchain[i]["transaction_count_all_time"]



        tmp_outer_dict = {"name": currencies[currency]["name"], "id": currencies[currency]["id"], "data": data_list}
        raw_data.append((tmp_outer_dict))
```

```python
f = open("raw_data.txt", "w")
f.write(str(raw_data))
f.close()
```
