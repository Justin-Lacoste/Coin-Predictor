```python
import ast
f = open('raw_data_2.txt')
data = ast.literal_eval(f.read())
print(len(data))
```

```python
list_of_parameters = ["time", "price", "volume_from", "volume_to", "comments", "points", "followers", "fb_likes", "twitter_followers", "twitter_statuses", "reddit_subscribers", "reddit_active_users", "reddit_posts_per_day", "reddit_comments_per_day", "new_addresses", "transaction_count", "large_transaction_count", "average_transaction_value", "unique_addresses_all_time", "current_supply", "block_time", "block_size", "hashrate", "block_height", "zero_balance_addresses_all_time", "transaction_count_all_time"]
#the following parameters are in more than 93% of days
list_of_parameters_93 = ["time", "price", "volume_from", "volume_to", "comments", "points", "followers", "fb_likes", "twitter_followers", "twitter_statuses", "reddit_subscribers", "reddit_active_users", "reddit_posts_per_day", "reddit_comments_per_day", "transaction_count", "large_transaction_count", "average_transaction_value", "unique_addresses_all_time", "current_supply"]
```

```python
#DATA OVERVIEW
total_days = 0
total_parameters = {}
for i in range(0, len(list_of_parameters)):
    total_parameters[list_of_parameters[i]] = 0

for i in range(0, len(data)):
    coin_data = data[i]["data"]
    for k in range(0, len(coin_data)):
        total_days += 1
        for j in range(len(total_parameters)):
            if (list_of_parameters[j] in coin_data[k]):
                if (list_of_parameters[j] == "price"):
                    if coin_data[k]["price"] != 0: total_parameters[list_of_parameters[j]] += 1
                else:
                    total_parameters[list_of_parameters[j]] += 1

for i in range(0, len(list_of_parameters)):
    total_parameters[list_of_parameters[i]] = (total_parameters[list_of_parameters[i]]/total_days)*100
print(total_days)
print(total_parameters)
    
    
```

```python
#create list of coins, exclusively those who have all the features who are in 93% of coins
list_of_coins_93 = [] #745

for i in range(0, len(data)):
    #creates empty object with parameters = 0
    total_days = 0
    total_parameters = {}
    for l in range(0, len(list_of_parameters_93)):
        total_parameters[list_of_parameters_93[l]] = 0

    coin_name = data[i]["name"]
    coin_data = data[i]["data"]
    for k in range(0, len(coin_data)):
        total_days += 1
        for j in range(len(total_parameters)):
            if (list_of_parameters_93[j] in coin_data[k]):
                if (list_of_parameters_93[j] == "price"):
                    if coin_data[999]["price"] > 0:
                        total_parameters[list_of_parameters_93[j]] += 1
                        if (coin_name == "0XBTC"):
                            print(coin_data[999]["price"])
                else:
                    total_parameters[list_of_parameters_93[j]] += 1
    areAllTrue = True
    
    
    for j in range(len(total_parameters)):
            if total_parameters[list_of_parameters_93[j]]-1 >= total_days:
                areAllTrue = False
    
    if areAllTrue:
        list_of_coins_93.append(coin_name)


print(len(list_of_coins_93))
```

```python
#structure:
#X = [[[day1],[day2],...,[day14]], [[...]]]
#Y = [increase_1, increase_2, ... incarease_n]
X_2_weeks_1_week = []
Y_2_weeks_1_week = []
```

```python
x_days = 14
y_days = 7
x_y_total = x_days + y_days

for i in range(0, len(data)):

    coin_name = data[i]["name"]
    coin_data = data[i]["data"]
    
    print(coin_name)
    
    hadNoDivisionByZero = True
    increase_for_y
    
    number_of_stacks = int(len(coin_data)/x_y_total)
    
    for stack_number in range(0, number_of_stacks):
        tmp_stack_array = []
        for item_in_stack_number in range(0, x_y_total):
            all_params_are_there = True
            for j in range(len(total_parameters)):
                if not (list_of_parameters_93[j] in coin_data[(stack_number*x_y_total)+item_in_stack_number]):
                    all_params_are_there = False
            if (all_params_are_there):
                tmp_array = []                        
                coin_data_for_the_day = coin_data[(stack_number*x_y_total)+item_in_stack_number]
                
                if (item_in_stack_number == x_days):
                    if coin_data_for_the_day["price"] != 0:
                        increase_for_y = (coin_data[(stack_number*x_y_total)+x_y_total-1]["price"]/coin_data_for_the_day["price"])-1                            
                    else:
                        hadNoDivisionByZero = False
                else:
                    for the_key, the_value in coin_data_for_the_day.items():
                        if (the_key != "time" and coin_name != "GARD" and coin_name != "POLY"):
                            tmp_array.append(float(the_value))
                    tmp_stack_array.append(tmp_array)
        if hadNoDivisionByZero and coin_name != "GARD" and coin_name != "POLY":
            Y_2_weeks_1_week.append(increase_for_y)
            X_2_weeks_1_week.append(tmp_stack_array)

        
print(len(X_2_weeks_1_week))
print(len(Y_2_weeks_1_week))
```
